## from collections import namedtuple

```python
from collections import namedtuple

Etudiant = namedtuple('Etudiant',['prenom','nom','age'])

stud_01 = Etudiant('Kévin','Silliau','20')

# Accès avec l'index
print(f"Le prénom de l'étudiant est {stud_01[0]}")

# Accès avec le nom de la variable
print(f"L'âge de l'étudiant est {stud_01.age}")
```

```python
stud_01._asdict()
```

```python
stud_01._fields
```

```python
stud_01 = Etudiant("Kévin","Parker", 20)
print(stud_01._replace(prenom="David")) # affiche Etudiant(prenom='David', age=20)
```

```python
Etudiant._make(stud_01)
```

## from collections import ChainMap
Permet d'encapsuler de nombreux dictionnaires dans une seule unité.

```python
from collections import ChainMap

fruits = {'pommes': 1, 'bananes': 2}
legumes = {'haricots': 3, 'carottes': 4}
feculents = {'pâtes': 5, 'riz': 6}

_chainmap = ChainMap(fruits, legumes, feculents)
_chainmap
```

```python
# Accéder à un élément spécifique à l'aide de sa clé
print(_chainmap['pommes'])  # Affiche : 1
print(_chainmap.get('bananes'))  # Affiche : 2

for key, value in _chainmap.items():
    print(key, "-->", value, end=" / ")
```

```python
# renvoi un ChainMap sans le premier dict
_chainmap.parents
```

```python
_chainmap.maps
```

```python
import os, argparse

defaults = {'color': 'red', 'user': 'guest'}

parser = argparse.ArgumentParser()
parser.add_argument('-u', '--user')
parser.add_argument('-c', '--color')

namespace = parser.parse_args()
command_line_args = {k: v for k, v in vars(namespace).items() if v is not None}

combined = ChainMap(command_line_args, os.environ, defaults)
print(combined['color'])
print(combined['user'])
```

## from collections import Counter

```python
from collections import Counter

_compteur = Counter() # Compteur vide
print(_compteur)

_compteur = Counter('Guido van Rossum') #compteur avec un itérable
print(_compteur)

_compteur = Counter({'Vert': 4, 'Bleu': 2}) # un compteur avec un mapping
print(_compteur)

_compteur = Counter(chats=4, chiens=7)#un compteur avec key=valeur
print(_compteur)

_compteur = Counter(["RED", "BLUE", "GREEN"])
print(_compteur)
print(_compteur['Yellow'])
```

```python
for e in _compteur.elements():
    print(e)
sorted(_compteur.elements())
```

## from collections import OrderedDict
depuis la version 3.7 les dictionnaires sont ordonné par défaut

```python
from collections import OrderedDict

mon_dict=OrderedDict()
mon_dict['pomme'] = '1'
mon_dict['banane'] = '2'
mon_dict['carotte'] = '3'
mon_dict['avocat'] = '4'
print(mon_dict)

mon_dict.move_to_end('banane')
print(mon_dict)

mon_dict.move_to_end('banane',last=False)
print(mon_dict)

print(mon_dict.popitem(True))

print(mon_dict)
```

## from collections import defaultdict

```python
from collections import defaultdict
mon_defaultdict = defaultdict(list)
print(mon_defaultdict["missing"]) # Affiche : []
```

```python
def default_message():
    return "No key for that, this is a default value for keys not defined"

mon_obj_defaultdict = defaultdict(default_message)
mon_obj_defaultdict["key1"] = "valeur1"
print(mon_obj_defaultdict["key1"]) # Affiche: valeur1
print(mon_obj_defaultdict["??"]) # Affiche: la clé n'existe pas
```
